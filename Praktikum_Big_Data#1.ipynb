```python
from pyspark.sql import SparkSession

# Memulai Spark session
spark = SparkSession.builder.appName("BigDataPractice").getOrCreate()

# Membuat DataFrame sederhana
data = [("Ali", 34, "Manajer", "Ngoding", "Laki-laki"), ("Budi", 23, "IT Manajer","Fotografi",'Laki-laki'), ("Citra", 29,"Dev ops", "Olahraga", "Perempuan"), ("Dina", 45,"IT Konsultan", "Bermain Game", "Perempuan")]
columns = ["Nama", "Usia", "Pekerjaan", "Hobi","Gender"]
df = spark.createDataFrame(data, columns)

# Menampilkan DataFrame
print ('Menampilkan data keseluruhan')
df.show()

# Filtering data
print ('Menampilkan data berdasarkan Usia > 25')
df_filtered = df.filter(df['Usia'] > 25)
df_filtered.show()

# Menghitung rata-rata usia
print ('Menampilkan rata-rata usia')
from pyspark.sql.functions import avg
df.groupBy().agg(avg("Usia")).show()

# Mengurutkan data berdasarkan usia
print ('Mengurutkan data berdasarkan Usia')
df_sorted = df.orderBy("Usia", ascending= True)
df_sorted.show()
```
